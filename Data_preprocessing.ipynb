```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler #ujednolicenie wartości, aby były z tego samego zakresu 
```

```python
import pymongo
mongodb = pymongo.MongoClient("mongodb://localhost:27017")
```

```python
db_estates = mongodb["db_estates"]
```

```python
X = []
Y = []
```

```python
kol_estates_x = db_estates["kol_estates_x"] 
kol_estates_y = db_estates["kol_estates_y"]
```

```python
for x in kol_estates_x.find():
    X.append(x)
    
for y in kol_estates_y.find():
    Y.append(y)
```

```python
X
```

```python
x = pd.DataFrame(X).drop(columns=["_id"])
```

```python
x
```

```python
y = pd.DataFrame(Y).drop(columns=["_id"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
scaler = MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train_scaled, y_train)
```

```python
x_train_scaled[0, :]
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

```python
len(linear_model.coef_[0, :])
```

```python
len(x_train_scaled[0, :])
```

```python
y_linear_model_train = linear_model.predict(x_train_scaled)
y_linear_model_train
```

```python
y_linear_model_test = linear_model.predict(x_test_scaled)
y_linear_model_test
```

```python
# mse, wizualizacja
def plot_results(y, y_pred, x_label, y_label, title, legend):
    plt.figure(figsize=(15, 10))
    plt.scatter(x=list(range(len(y_pred))), y=y_pred, c="r", alpha=0.2)
    
    plt.scatter(x=list(range(len(y))), y=y, c="g",alpha=0.2)
    
    plt.title(title)
    plt.legend(legend)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    
    
    
```

```python
y_train.max()
```

```python
plot_results(y_train, y_linear_model_train, "Nr of samlpes", "Price", title="Prdeciting prices - train", legend=["prediction", "real_data"] )
```

```python
plot_results(y_test, y_linear_model_test, "Nr of samlpes", "Price", title="Prdeciting prices - train", legend=["prediction_test", "real_data"] )
```

```python
# Root Mean Square Error (RMSE)
# suma kwadratów różnic wobec wartości n, pierwiastek z wariancji
# RMSE pokaże nam rzeczywiste ceny

# Przeuczenie modelu - na wykresie kropki dla danych treningowych pokrywają się z tymi, które występują z danymi rzeczywistymi
# przeuczenie - dla danych treningowych pokryte, dla testowych duże rozbieżności. Ogranicz parametry modelu.
# w przypadku sieci neuronowych zmniejsz liczbę epok

from sklearn.metrics import mean_squared_error


print(f"Regression train data RMSE: {mean_squared_error(y_train, y_linear_model_train, squared=False)}")
print(f"Regression test data RMSE: {mean_squared_error(y_test, y_linear_model_test, squared=False)}")

```

```python
def get_score(model, x_train, y_train, x_test, y_test):
    model.fit(x_train, y_train)
    return model.score(x_test, y_test)
    
```
