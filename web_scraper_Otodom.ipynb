# MAIN LOOP

```python
import traceback

#get data from pages
#( 1 to nr_of_pages)
# Get data from pages (1: number of pages)
for page in range(2, (nr_of_pages+1)):
    print(page)
    
    try:
        # Get not promoted adds
        not_promoted = soup.find_all('ul', {'class': 'css-14cy79a e3x1uf06'})[1]
        
        # Adds URLs
        hrefs = []
        # Fill hrefs list
        for offer in not_promoted:
            try:
                # Get href and create a full URL
                href = offer.find_all("a", {"class": "css-1c4ocg7 es62z2j23"})[0]['href']
                hrefs.append('https://www.otodom.pl' + href)
            except:
                print('HREF')
                traceback.print_exc()
                
        # Create lists for data and column storing
        col_names_to_add = ['Tytul', 'Cena', 'Lokalizacja', 'URL']
        data_to_add = []
        
        column_names = []
        data_csv = []
        
        # Iterate through article hrefs
        for article_no, article_href in enumerate(hrefs):
            try:
                print("Page: " + str(page-1) + '/' + str(nr_of_pages) + ', add: ' + str(article_no+1) + '/' + str(len(hrefs)))
                time.sleep(0.3)
                r_offer = requests.get(article_href)
                soup_offer = bs(r_offer.content, features = "html.parser") 

                offer_title = soup_offer.find("h1", {"class": "css-11kn46p eu6swcv17"}).get_text()


                offer_price = soup_offer.find("strong", {"class": "css-8qi9av eu6swcv16"}).get_text()
                offer_price = offer_price.replace(" ", "")
                offer_price = offer_price.split("zł")[0]

                offer_address = soup_offer.find("a", {"class": "e1nbpvi60 css-1kforri e1enecw71"}).get_text()

                data_to_add.extend([offer_title, offer_price, offer_address, article_href])

                parameters_class = soup_offer.find("div", {"class": "css-wj4wb2 emxfhao1"})

                parameters = parameters_class.find_all("div", {"class": "css-1ccovha estckra9"})

                param_col_name = []
                param_data =[]

                for param in parameters:
                    param_col_name.append(param.select("div")[1].get_text())
                    param_data.append(param.select("div")[2].get_text())

                additional_parameters_class = soup_offer.find("div", {"class": "css-1l1r91c emxfhao1"})

                if additional_parameters_class:
                    additional_parameters = additional_parameters_class.find_all("div", {"class": "css-f45csg estckra9"})

                    for additional_param in additional_parameters:
                        param_col_name.append(additional_param.select("div")[1].get_text())
                        param_data.append(additional_param.select("div")[2].get_text())

                    col_names_to_add.extend(param_col_name)
                    data_to_add.extend(param_data)

                    column_names.append(col_names_to_add)
                    data_csv.append(data_to_add)

                    col_names_to_add = ['Tytul', 'Cena', 'Lokalizacja', 'URL']
                    data_to_add = []
                
            except:
                traceback.print_exc()
                print("Error! Page: " + str(page-1) + 'add: ' + str(article_no))

                # Clear lists for current add (error)
                col_names_to_add = ['Tytul', 'Cena', 'Lokalizacja', 'URL']
                data_to_add = []
        
        for index in range(len(data_csv)):
            df = df.append(pd.DataFrame([data_csv[index]], columns = column_names[index]), sort=False)
            
        
    except:
        print("Error FINAL! Page: " + str(page-1))
        traceback.print_exc()
    

    # URL update
    link = link_base + str(page)
    # Make a request to first page
    r = requests.get(link)
    # Create a soup object
    soup = bs(r.content, features = "html.parser")
    
    print(link)
```

```python
# Save data to csv file
filepath = f'data_scraping.csv'
df.to_csv(filepath, sep=';', index=False, encoding='utf-8-sig')
```
